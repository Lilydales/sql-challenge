```python
import pandas as pd
import os
```

```python
#Dependancies
folder='data'
#import data
departments=pd.read_csv(os.path.join(folder,'departments.csv'))
dept_emp=pd.read_csv(os.path.join(folder,'dept_emp.csv'))
dept_manager=pd.read_csv(os.path.join(folder,'dept_manager.csv'))
employees=pd.read_csv(os.path.join(folder,'employees.csv'))
salaries=pd.read_csv(os.path.join(folder,'salaries.csv'))
titles=pd.read_csv(os.path.join(folder,'titles.csv'))
```

```python
#List the employee number, last name, first name, sex, and salary of each employee.
employee_salaries=pd.merge(employees,salaries,on="emp_no").loc[:,['emp_no','last_name',
                                                                 'first_name','sex','salary']]
employee_salaries.head()
```

```python
# Convert 'hire_date' to datetime format
employees['hire_date'] = pd.to_datetime(employees['hire_date'])

#List the first name, last name, and hire date for the employees who were hired in 1986.
hired_in_1986=employees[['first_name','last_name','hire_date']].loc[employees['hire_date'].dt.year==1986]

hired_in_1986.head()
```

```python
#List the manager of each department along 
#with their department number, department name, employee number, last name, and first name.
managers_details=pd.merge(dept_manager,departments, on='dept_no')
managers_details=pd.merge(managers_details,employees,on='emp_no').loc[:,['dept_no','dept_name',
                                                                  'emp_no','last_name',
                                                                  'first_name']]
managers_details.head()
```

```python
#List the department number for each employee along with 
#that employee’s employee number, last name, first name, and department name.
employees_details=pd.merge(dept_emp,departments,on='dept_no')
employees_details=pd.merge(employees_details,employees,on='emp_no').loc[:,['dept_no','dept_name',
                                                                          'emp_no','last_name',
                                                                          'first_name']]
employees_details.head()
```

```python
#List the first name, last name, and sex of each employee 
#whose first name is Hercules and whose last name begins with the letter B.
employees_Hercules_B=employees.loc[(employees['first_name']=='Hercules') & (
    employees['last_name'].str.startswith('B')),['first_name','last_name','sex']]
employees_Hercules_B.head()
```

```python
#List each employee in the Sales department,
#including their employee number, last name, and first name.
employees_sales=pd.merge(dept_emp,departments,on='dept_no')
employees_sales=pd.merge(employees_sales,employees,on='emp_no').loc[employees_sales['dept_name']=='Sales',[
    'emp_no','first_name','last_name']]
employees_sales.head()
```

```python
#List each employee in the Sales or Development department,
#including their employee number, last name, and first name.
employees_sales=pd.merge(dept_emp,departments,on='dept_no')
employees_sales=pd.merge(employees_sales,employees,
                         on='emp_no').loc[employees_sales['dept_name'].isin(['Sales', 'Development']),[
    'emp_no','first_name','last_name','dept_name']]
employees_sales.head()
```

```python
#List the frequency counts, in descending order, 
#of all the employee last names (that is, how many employees share each last name).
employee_lastname_shared=employees.groupby(
    'last_name').size().reset_index(name='frequency counts').sort_values(
    by='frequency counts',ascending=False)
employee_lastname_shared.head()
```
